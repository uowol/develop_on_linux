```python
import requests
from bs4 import BeautifulSoup as bs
import numpy as np
import pandas as pd
```

```python
import re

def extract_words(title):
    # words = re.findall('[0-9]{1,3},[0-9]{1,3}[ㄱ-횧0-9a-z\.%]+|[ㄱ-횧0-9a-z\.%]+', title)
    words = re.findall('[ㄱ-횧a-zA-Z]+', title)   # without 수치, 날짜
    return words

    # fixed_words = []
    # for word in words:
    #     fixed_words.extend(re.findall(".*[^의는을를부터합니다하겠습니다까지또는및등한다에]", word))
    # return fixed_words
```

```python
words_list = []

print("crawling...")
for page_index in range(308):
    url = "http://ncov.mohw.go.kr/tcmBoardList.do?pageIndex={}&brdId=&brdGubun=&board_id=&search_item=1&search_content=".format(str(page_index))

    # print(url)
    response = requests.get(url)

    if response.status_code == 200:
        html = response.text
        soup = bs(html, 'html.parser')
        titles = [a.get_text() for a in soup.select('#content > div > div.board_list > table > tbody > tr > td.ta_l > a')]
        # print(titles)

        words_list.extend([extract_words(title) for title in titles])
    else:
        print(response.status_code)

print("end!")

```

```python
dict_words_list = []
dict_words_global = {}
keywords = set()

for words in words_list:
    dict_words = {}
    for word in words:
        dict_words.setdefault(word, 0)
        dict_words_global.setdefault(word, 0)
        dict_words[word] += 1
        dict_words_global[word] += 1
        keywords.add(word)
    dict_words_list.append(dict_words)

dict_words_global
```

```python
df = pd.DataFrame(data=[[0]*len(keywords)]*len(dict_words_list), index=range(0,len(dict_words_list)), columns=sorted(keywords))
```

```python
for i, dict_words in enumerate(dict_words_list):
    for keyword, value in dict_words.items():
        df.loc[i,keyword] = value
df
```

```python
df.to_csv("withCorona.csv")
```
